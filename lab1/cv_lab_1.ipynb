```python
from matplotlib import image
import matplotlib.pyplot as plt
import numpy as np

import os

IMAGES_PATH = os.getcwd() + "/images/"
K = [1, 0.75, 0.5, 0.25]
L = [0, 25, 50, 75]
```

```python
def linear_filtering(fname, k=1., l=0) -> np.ndarray:
    img = image.imread(fname=IMAGES_PATH + fname)
    img = img.astype(float)
    img = img*k + l
    img[img > 255] = 255
    img[img < 0] = 0
    return img.astype(int)
```

```python
def output(fname):
    fig, axs = plt.subplots(4, 4, figsize=(50,25))
    for i in range(4):
        for j in range(4):
            axs[i, j].set_axis_off()
            axs[i, j].set_title(f"{fname.rsplit('.')[0]} : K = {K[i]}, L = {L[j]}", fontsize=25)
            axs[i, j].imshow(linear_filtering(fname, K[i], L[j]))
```

```python
for fname in os.listdir(IMAGES_PATH):
    output(fname)
```
